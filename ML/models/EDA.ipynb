```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sb
```

```python
# Data Collection and inital exploration 
data = pd.read_csv("../datasets/house-price/Bangalore.csv")
print(data.head())
```

```python
print(f"Data Shape: {data.shape}\n")
print(f"Data information:\n {data.info()}\n")
```

```python
# check for missing values 
print(f"Missing values in each column:\n {data.isnull().sum()}\n")
```

```python
print(f"Statistical summary of numerical features:\n {data.describe()}\n")
```

```python
# Single/Target variable analysis
plt.figure()
data["Price"].hist(bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# scatter plot for area and price
plt.figure()
sb.scatterplot(x=data["Area"], y=data["Price"], hue=data['Area'] , palette='viridis' )
plt.xlabel("Area (sqft)")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
# Bar plot for average price by number of bedrooms
data.groupby("No. of Bedrooms")["Price"].mean().plot(kind="bar")
plt.title("Average Price by Number of Bedrooms")
plt.ylabel("Average Price")
plt.show()
```

```python
# Top 10 locations by number of listings
top_locations = data['Location'].value_counts().head(10)
top_locations
data.groupby("Location")["Price"].mean().sort_values(ascending=False).head(10)
```

```python
# resale impact on price
print(f"Average Price for Resale Properties: \n")
data.groupby("Resale")["Price"].mean()
```

```python
plt.figure()
plt.boxplot(data["Price"])
plt.title("Price Boxplot")
plt.show()

plt.figure()
plt.boxplot(data["Area"])
plt.title("Area Boxplot")
plt.show()
```
